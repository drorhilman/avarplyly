```python
from glob import glob
import os
import pandas as pd
import feedparser


NewsFeed = feedparser.parse("https://feed.podbean.com/plyly/feed.xml")
entries = pd.DataFrame(NewsFeed['entries'])
entries
```

```python
e = entries[['title', 'link', 'published', 'id', 'itunes_duration']].copy()
e['published'] = pd.to_datetime(e['published'])
e.style
```
